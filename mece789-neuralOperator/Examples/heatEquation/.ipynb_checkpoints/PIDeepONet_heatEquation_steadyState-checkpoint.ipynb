## Steady-State Heat Equation -- PIDeepONet

PDE:
$$-\nabla^2 T(x,y) = q(x,y), \quad (x,y)\in\Omega=[0,1]^2$$

### (1.1) Boundary conditions

| Wall | Condition |
|---|---|
| Left $x=0$ | $T = 1$ (hot wall) |
| Right $x=1$ | $T = 0$ (cold wall) |
| Top/bottom $y=0,1$ | $\partial T/\partial n = 0$ (insulated, Neumann) |

### (1.2) Neural operator learning problem

$$\mathcal{G}:\; q(x,y) \;\longrightarrow\; T(x,y)$$

**q-field:** $q(x,y) = e^{\alpha x + \beta y}$, $\alpha,\beta \sim \mathrm{Uniform}(-1,1)$

**Mollifier** -- mirrors the professor's mass diffusion problem exactly:
$$T = u_{\mathrm{NN}}(x,y)\cdot x(1-x)\cdot\sin(\pi y) + (1-x)$$

Verification:
- $x=0$: $T = 0 + 1 = 1$ (hot wall) $\checkmark$
- $x=1$: $T = 0 + 0 = 0$ (cold wall) $\checkmark$
- $y=0,1$: $\sin(\pi y)=0 \Rightarrow T = 1-x$ -- linear profile consistent with zero normal flux $\checkmark$

The particular solution $(1-x)$ anchors the network output to the correct scale
from random initialisation, so pure PI training ($w_{\mathrm{data}}=0$) works --
exactly as in the professor's mass diffusion example.

## (2) The Physics-Informed DeepONet

```python
import sys
sys.path.append('../..')
import numpy as np
import h5py
import torch
import matplotlib.pyplot as plt

def setup_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    torch.backends.cudnn.deterministic = True
random_seed = 1234
setup_seed(random_seed)

if torch.cuda.is_available():
    device = 'cuda:0'
elif torch.backends.mps.is_available():
    device = 'mps'
else:
    device = 'cpu'

dtype = torch.float32
tag   = 'heat_exp_nonzeroBC'
res   = 29

######################################
# Load dataset
######################################
data_train = h5py.File('../../Problems/heatEquation_2d/heat_exp_nonzeroBC_train.mat', 'r')
data_test  = h5py.File('../../Problems/heatEquation_2d/heat_exp_nonzeroBC_test.mat',  'r')
print('train keys:', list(data_train.keys()))
print('test  keys:', list(data_test.keys()))

from Utils.utils import *
n_train, n_test = 1000, 200

def get_data(data, ndata, dtype, n0=0):
    a = np2tensor(np.array(data['coeff'][..., n0:n0+ndata]), dtype)
    u = np2tensor(np.array(data['sol_fdm'][..., n0:n0+ndata]), dtype)
    X = np.array(data['X']).T
    Y = np.array(data['Y']).T
    mesh  = np2tensor(np.vstack([X.ravel(), Y.ravel()]).T, dtype)
    gridx = mesh.reshape(-1, 2)
    x     = gridx.repeat((ndata, 1, 1))
    a     = a.reshape(ndata, -1, 1)
    u     = u.reshape(ndata, -1, 1)
    return a, u, x, gridx

a_train, u_train, x_train, grid_train = get_data(data_train, n_train, dtype)
a_test,  u_test,  x_test,  grid_test  = get_data(data_test,  n_test,  dtype)
print('a_train:', a_train.shape, '  u_train:', u_train.shape)
print('x_train:', x_train.shape, '  grid_train:', grid_train.shape)

######################################
# Structured mesh for FDM residual
######################################
from Utils.GenPoints import Point2D
pointGen = Point2D(x_lb=[0., 0.], x_ub=[1., 1.], dataType=dtype, random_seed=random_seed)
N_mesh = 29
h      = 1.0 / (N_mesh - 1)
x_mesh = pointGen.inner_point(N_mesh, method='mesh')  # (N_mesh^2, 2)
print('x_mesh shape:', x_mesh.shape)


print("x_mesh min/max x:", x_mesh[:,0].min().item(), x_mesh[:,0].max().item())
print("x_mesh min/max y:", x_mesh[:,1].min().item(), x_mesh[:,1].max().item())
print("x_mesh shape:", x_mesh.shape)
print("First 5 points:", x_mesh[:5])
print("Last  5 points:", x_mesh[-5:])
```

### (3.2) Define the loss class and train the model

```python
import torch.nn as nn
from torch.autograd import Variable

########################################
class fun_q(object):
    # Nearest-grid-point lookup of q(x,y).
    # Flat index: iy*res + ix  (C-order).

    def __init__(self, res):
        self.res   = res
        self.delta = 1.0 / (res - 1)

    def __call__(self, x, a):
        a     = a.squeeze(-1)
        x_loc = torch.floor(x[..., 0] / self.delta + 0.5).int().clamp(0, self.res-1)
        y_loc = torch.floor(x[..., 1] / self.delta + 0.5).int().clamp(0, self.res-1)
        loc   = y_loc * self.res + x_loc
        img   = a[torch.arange(a.size(0)).unsqueeze(1), loc]
        return img.unsqueeze(-1)

########################################
class mollifier(object):
    # Mirrors the professor's mass diffusion mollifier exactly:
    #   T = u_NN * x*(1-x)*sin(pi*y) + (1-x)
    #
    # The (1-x) particular solution means:
    #   - At random init, T ~ 1-x (correct scale immediately)
    #   - Network only learns the perturbation from q
    #   - Pure PI training (w_data=0) works, same as mass diffusion

    def __call__(self, u, x):
        xx = x[..., 0:1]
        yy = x[..., 1:2]
        return u * xx * (1.0 - xx) * torch.sin(np.pi * yy) + (1.0 - xx)

################################
class LossClass(object):

    def __init__(self, solver):
        self.solver    = solver
        self.dtype     = solver.dtype
        self.device    = solver.device
        self.fun_q     = fun_q(res=res)
        self.model_u   = solver.model_dict['u']
        self.mollifier = mollifier()

    def Loss_pde(self, a_batch, w_pde):
        # PDE residual via direct 5-point Laplacian stencil.
        #
        # T = u_NN * x*(1-x)*sin(pi*y) + (1-x)
        # Lap(T) = Lap(u_NN * x*(1-x)*sin(pi*y))  +  Lap(1-x)
        #        = Lap(mollifier term)              +  0
        # Residual: Lap(T) + q = 0
        #
        # dim convention: dim1 = y (rows), dim2 = x (cols)
        n_batch = a_batch.shape[0]
        if w_pde > 0.0:
            x = Variable(
                x_mesh.repeat(n_batch, 1, 1).to(self.device),
                requires_grad=False
            )

            # Network prediction + mollifier
            T = self.model_u(x, a_batch)
            T = self.mollifier(T, x).reshape(-1, N_mesh, N_mesh, 1)

            # Direct 5-point Laplacian stencil
            T_c = T[:, 1:-1, 1:-1, :]   # center
            T_l = T[:, 1:-1, :-2,  :]   # x-left
            T_r = T[:, 1:-1, 2:,   :]   # x-right
            T_d = T[:, :-2,  1:-1, :]   # y-down
            T_u = T[:, 2:,   1:-1, :]   # y-up
            laplacian = (T_l + T_r + T_d + T_u - 4.0 * T_c) / (h ** 2)

            # q at interior points
            q_col = self.fun_q(x, a_batch).reshape(-1, N_mesh, N_mesh, 1)
            q_col = q_col[:, 1:-1, 1:-1, :]

            # Residual: Lap(T) + q = 0
            # Normalise residual by h^2 so it's O(1) not O(h^-2)
            residual = (laplacian + q_col).reshape(n_batch, -1) * (h ** 2)
            return self.solver.getLoss(residual, torch.zeros_like(residual))
        return torch.tensor(0.0)

    def Loss_data(self, x, a, u, w_data):
        if w_data > 0.0:
            u_pred = self.model_u(x, a)
            u_pred = self.mollifier(u_pred, x)
            return self.solver.getLoss(u_pred, u)
        return torch.tensor(0.0)

    def Error(self, x, a, u):
        u_pred = self.model_u(x, a)
        u_pred = self.mollifier(u_pred, x)
        return self.solver.getError(u_pred, u)

######################################
# Model setup
######################################
from Solvers.PIDeepONet import PIDeepONet
solver  = PIDeepONet.Solver(device=device, dtype=dtype)
netType = 'DeepONetBatch'

from Networks.CNNet import CNNet2d

class BranchNet(nn.Module):
    def __init__(self, conv_arch, fc_arch, nx_size, ny_size, dtype=None):
        super().__init__()
        self.nx_size = nx_size
        self.ny_size = ny_size
        self.conv = CNNet2d(
            conv_arch=conv_arch, fc_arch=fc_arch,
            activation_conv='SiLU', activation_fc='SiLU',
            kernel_size=(3, 3), stride=2, dtype=dtype
        )
    def forward(self, x):
        x = x.reshape(-1, self.ny_size, self.nx_size).unsqueeze(1)
        return self.conv(x)

conv_arch = [1, 64, 64, 64]
fc_arch   = [64*2*2, 128, 128, 128]
branchNet = BranchNet(conv_arch, fc_arch, nx_size=res, ny_size=res, dtype=dtype)

layers_branch, activation_branch = [branchNet, fc_arch[-1]], 'SiLU'
layers_trunk,  activation_trunk  = [2, 128, 128, 128, 128], 'ReLU'
model_u = solver.getModel(
    layers_branch, layers_trunk, activation_branch, activation_trunk,
    multi_ouput_strategy=None, num_output=1, netType=netType
)
n_params = sum(p.numel() for p in model_u.parameters() if p.requires_grad)
print(f'{n_params:,} trainable parameters.')
```

### (2.3) Training and prediction

#### (2.3.1) Train the model

```python
print("u_test range:", u_test.min().item(), u_test.max().item())
print("u_train range:", u_train.min().item(), u_train.max().item())
print("a_test range:", a_test.min().item(), a_test.max().item())

# Manually compute what getError should give
with torch.no_grad():
    u_pred_check = model_u(x_test[:5].to(device), a_test[:5].to(device))
    u_pred_check = mollifier()(u_pred_check, x_test[:5].to(device)).cpu()

print("u_pred shape:", u_pred_check.shape)
print("u_pred range:", u_pred_check.min().item(), u_pred_check.max().item())
print("u_test[0:5] range:", u_test[:5].min().item(), u_test[:5].max().item())

# Manual relative L2
err_manual = torch.norm(u_pred_check - u_test[:5]) / torch.norm(u_test[:5])
print("Manual relative L2 (5 samples):", err_manual.item())
print("solver.getError (5 samples):", solver.getError(u_pred_check, u_test[:5]))



# Quick check -- u_test should NOT be constant
print("u_test sample 0, first 10 values:", u_test[0, :10, 0])
print("u_test sample 0, last  10 values:", u_test[0, -10:, 0])


raw = np.array(data_test['sol_fdm'])
print("sol_fdm raw shape:", raw.shape)
print("sol_fdm raw[0,0,:]:", raw[0,0,:5])   # first row, first col, first 5 samples
print("sol_fdm raw[:,0,0]:", raw[:5,0,0])   # first 5 rows, first col, first sample
print("sol_fdm raw[0,:,0]:", raw[0,:5,0])   # first row, first 5 cols, first sample

print("a_test sample 0, first 10 values:", a_test[0, :10, 0])
```

```python
model_dict = {'u': model_u}
solver.train_setup(model_dict, lr=1e-3, optimizer='Adam', scheduler_type='StepLR',
                   gamma=0.6, step_size=200)
solver.train(LossClass, a_train, u_train, x_train, a_test, u_test, x_test,
             w_data=1., w_pde=1., batch_size=50, epochs=2000, epoch_show=100,
             **{'save_path': f'saved_models/PI{netType}_{tag}/'})
```

### (3.3) Load saved model and make prediction

```python
#######################################
# Load the trained model
#######################################
from Solvers.PIDeepONet import PIDeepONet
from torch.autograd import Variable

solver = PIDeepONet.Solver(device=device, dtype=dtype)
model_trained = solver.loadModel(
    path=f'saved_models/PI{netType}_{tag}/',
    name='model_pideeponet_besterror'
)

#########################################
# Predict on test set
#########################################
x_var = Variable(x_test.to(device), requires_grad=False)
a_var = a_test.to(device)
with torch.no_grad():
    u_pred = model_trained['u'](x_var, a_var)
    u_pred = mollifier()(u_pred, x_var).detach().cpu()

print('a_test  shape:', a_test.shape)
print('u_test  shape:', u_test.shape)
print('u_pred  shape:', u_pred.shape)
print('Test L2 error (avg):', solver.getError(u_pred, u_test))

#########################################
# Visualise one sample
#########################################
inx = 0
from Utils.PlotFigure import Plot
Plot.show_2d_list(
    [grid_test] * 4,
    [a_test[inx], u_test[inx], u_pred[inx], torch.abs(u_test[inx] - u_pred[inx])],
    ['q (heat source)', 'T_true', 'T_pred', '|T_true - T_pred|'],
    lb=0., ub=1.
)

#########################################
# Loss and error curves
#########################################
loss_saved = solver.loadLoss(
    path=f'saved_models/PI{netType}_{tag}/', name='loss_pideeponet'
)
Plot.show_loss(
    [loss_saved['loss_train'], loss_saved['loss_test'], loss_saved['loss_pde']],
    ['loss_train', 'loss_test', 'loss_pde']
)
Plot.show_error([loss_saved['time']], [loss_saved['error']], ['l2_test T'])
```
